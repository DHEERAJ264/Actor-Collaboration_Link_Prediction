```python
import numpy as np
import matplotlib.pyplot as plt

# data to plot
n_groups = 4
precision = (62,78.7,69.8,79.2)
recall = (33.5,46.7,40.8,46.6)

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index,precision, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Precision')
 
rects2 = plt.bar(index + bar_width, recall, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Recall')
 
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('PR values of Various Algorithms')
plt.xticks(index + bar_width,('Jaccard-Coefficient','Adamic/Adar','Common Neighbors','Resource-Allocation'),rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
